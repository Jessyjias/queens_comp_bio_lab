1. Label each sample the mutations it has, the driver labels and the associated signatures with exposure.

```python
# read signatures from mutagene identify function 
# sig_txt = pd.read_csv('test_output_signature_identify_49')
# sig_txt.to_csv('test_output_signature_identify_49.csv')

##### NEED TO MANUALLY SEPARATE THE COLS IN EXCEL, renamed to signatures.csv

signatures = pd.read_csv('signatures_COAD.csv')
signatures
```

```python
# Add in the one hot encoding ver of signatures 
def merge_signature_exposure_by_sample(df, df_sig, signatures): 
    'signatures = array of signatures present in df_sig' 
    temp_df = df.reset_index(drop=True)
    temp_sigs = df_sig.reset_index(drop=True)
    
    for sig_col in signatures: 
        temp_df[sig_col] = pd.Series(dtype=temp_sigs['exposure'].dtype)

    for index, row in temp_df.iterrows():
        indexes_samp = temp_sigs.index.values[row['Tumor_Sample_Barcode']==temp_sigs['sample']]
        for ind in indexes_samp: 
            sig = temp_sigs.iloc[ind]['signature']
            # print(ind)
            # print(sig)
            # print(temp_sigs.iloc[index_samp]['exposure'])
            temp_df.at[index, sig] = temp_sigs.iloc[ind]['exposure']

    return temp_df
```

```python
df.ExpPOLEDriverLabel.unique()
```

```python
## Add another column to indicate both DXMY
## D = ['No mutation', 'Non driver', 'Driver'] = [0, 1, 2]
## M = ['MSS', 'MSI-L', 'MSI-H', 'nan] = [0, 1, 2, 3]

if 'POLEDriverMSIStatus' in df:
    df = df.drop(columns=["POLEDriverMSIStatus"])
if 'POLEDriverMSIStatus' not in df: 
    df['POLEDriverMSIStatus'] = pd.Series(dtype="string")
for index, row in df.iterrows():
    # no mutatin cases 
    if row['ExpPOLEDriverLabel'] == 'No mutation' and row['MSI'] == 'MSS': 
        df.at[index, 'POLEDriverMSIStatus'] = 'D0M0'
    elif row['ExpPOLEDriverLabel'] == 'No mutation' and row['MSI'] == 'MSI-L': 
        df.at[index, 'POLEDriverMSIStatus'] = 'D0M1'
    elif row['ExpPOLEDriverLabel'] == 'No mutation' and row['MSI'] == 'MSI-H': 
        df.at[index, 'POLEDriverMSIStatus'] = 'D0M2'
    elif row['ExpPOLEDriverLabel'] == 'No mutation' and row['MSI'] == np.nan: 
        df.at[index, 'POLEDriverMSIStatus'] = 'D0M3'
    # non driver mutation cases
    elif row['ExpPOLEDriverLabel'] == 'Non driver' and row['MSI'] == 'MSS': 
        df.at[index, 'POLEDriverMSIStatus'] = 'D1M0'
    elif row['ExpPOLEDriverLabel'] == 'Non driver' and row['MSI'] == 'MSI-L': 
        df.at[index, 'POLEDriverMSIStatus'] = 'D1M1'
    elif row['ExpPOLEDriverLabel'] == 'Non driver' and row['MSI'] == 'MSI-H': 
        df.at[index, 'POLEDriverMSIStatus'] = 'D1M2'
    elif row['ExpPOLEDriverLabel'] == 'Non driver' and row['MSI'] == np.nan: 
        df.at[index, 'POLEDriverMSIStatus'] = 'D1M3'
    # driver mutation cases
    elif row['ExpPOLEDriverLabel'] == 'Driver' and row['MSI'] == 'MSS': 
        df.at[index, 'POLEDriverMSIStatus'] = 'D2M0'
    elif row['ExpPOLEDriverLabel'] == 'Driver' and row['MSI'] == 'MSI-L': 
        df.at[index, 'POLEDriverMSIStatus'] = 'D2M1'
    elif row['ExpPOLEDriverLabel'] == 'Driver' and row['MSI'] == 'MSI-H': 
        df.at[index, 'POLEDriverMSIStatus'] = 'D2M2'
    elif row['ExpPOLEDriverLabel'] == 'Driver' and row['MSI'] == np.nan: 
        df.at[index, 'POLEDriverMSIStatus'] = 'D2M3'
    else: 
        df.at[index, 'POLEDriverMSIStatus'] = 'error'
```
